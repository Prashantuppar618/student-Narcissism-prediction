```python
#FILLING NULL VALUES
mode_value = test['GADE'].mode()[0]
test['GADE'].fillna(mode_value, inplace=True)
mode_value = test['League'].mode()[0]  
test['League'].fillna(mode_value, inplace=True)
mean_value = test['streams'].mean()
test['streams'] = test['streams'].fillna(mean_value)
mode_value = test['Narcissism'].mode()[0]
test['Narcissism'].fillna(mode_value, inplace=True)
mode_value = test['Work'].mode()[0]
test['Work'].fillna(mode_value, inplace=True)
mode_value = test['Degree'].mode()[0]
test['Degree'].fillna(mode_value, inplace=True)
mode_value = test['Reference'].mode()[0]
test['Reference'].fillna(mode_value, inplace=True)
mode_value = test['accept'].mode()[0]
test['accept'].fillna(mode_value, inplace=True)
median_value = test['SPIN_T'].median()
test['SPIN_T'] = test['SPIN_T'].fillna(median_value)
mean_value = test['Hours'].mean()
test['Hours'] = test['Hours'].fillna(mean_value)
mode_value = test['Residence_ISO3'].mode()[0]  
test['Residence_ISO3'].fillna(mode_value, inplace=True)
mode_value = test['Birthplace_ISO3'].mode()[0]  
test['Birthplace_ISO3'].fillna(mode_value, inplace=True)
```

```python
# FINDING THE NUMBER OF OUTLIERS OF EACH COLUMN
numerical_data = test.select_dtypes(include=['float64', 'int64'])  # Select numerical attributes

# Calculate the first and third quartiles (Q1 and Q3)
Q1 = numerical_data.quantile(0.25)
Q3 = numerical_data.quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define a threshold for outliers (e.g., 1.5 times the IQR)
threshold = 1.5

# Identify outliers
outliers = (numerical_data < (Q1 - threshold * IQR)) | (numerical_data > (Q3 + threshold * IQR))

# Print the number of outliers for each attribute
print(outliers.sum())
```

```python
#DROPING OUTLIERES
test_no_outliers=test[~outliers]
```

## FEATURE ENGINEERING

```python
le = LabelEncoder()
for i in test.columns:
    if test[i].dtype == 'object':
        test[i] = le.fit_transform(test[i])
```

```python
from sklearn.decomposition import PCA
df1 = test[['GAD_T'  , 'SWL_T' , 'SPIN_T' ]]
df2 = test[[ 'Age' ,  'Hours' , 'streams' ]]
pc1 = PCA(n_components=2)
pc2 = PCA(n_components=2)
x1 = pc1.fit_transform(df1)
x2 = pc2.fit_transform(df2)
x = x1 + x2
```

```python
from sklearn.cluster import KMeans
WCSS = []
for i in range(1,12):
    model = KMeans(n_clusters = i,init = 'k-means++')
    model.fit(x)
    WCSS.append(model.inertia_) #inertia --> error
fig = plt.figure(figsize = (7,7))
plt.plot(range(1,12),WCSS, linewidth=4, markersize=12,marker='o',color = 'Hotpink')
plt.xticks(np.arange(12))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters = 5, init = "k-means++", max_iter = 300, n_init = 40, random_state = 0)
y_clusters = model.fit_predict(x)
pd.Series(y_clusters).value_counts().plot(kind='bar', color='Gold');
```

```python
plt.scatter(x[y_clusters == 0, 0], x[y_clusters == 0, 1], s = 60, c = 'orange', label = 'Cluster1')
plt.scatter(x[y_clusters == 1, 0], x[y_clusters == 1, 1], s = 60, c = 'red', label = 'Cluster2')
plt.scatter(x[y_clusters == 2, 0], x[y_clusters == 2, 1], s = 60, c = 'blue', label = 'Cluster3')
plt.scatter(x[y_clusters == 3, 0], x[y_clusters == 3, 1], s = 60, c = 'violet', label = 'Cluster4')
plt.scatter(x[y_clusters == 4, 0], x[y_clusters == 4, 1], s = 60, c = 'yellow', label = 'Cluster5') 

plt.legend()
plt.show()
```
